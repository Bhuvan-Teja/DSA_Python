### Bubble sort
- Bubble sort uses the logic of taking 1st element and comparing the next the element if the 1st element is bigger then swap
- which leads to the pushing the max number to the last
- in each itteration you will push the max to the last and shrink the size or array from right.

```python
class Solution():
    def bubble_sort(self, nums):
        n = len(nums)
        for i in range(n): # 0 to n-1, for n = 5, 0 --> 4
            for j in range(n-i-1): # 0 to n-i-1, for n=5, 1--> 3
                if nums[j] >= nums[j+1]:
                    nums[j], nums[j+1] = nums[j+1], nums[j]
        return nums
```

```python
sol = Solution()
ar = [7,4,1,5,3,7]
sol.bubble_sort(ar)
print(ar)
```

#### Bubble using Recursion
- Using Recursion we can solve for bubble sort
- for each ittiration process the sorting and decrease the length of array by 1 at the end

```python
class Solution_R():

    def bubble_sort_helper(self, nums, n):
        if n == 1:
            return
        for i in range(n-1):
            if nums[i]>nums[i+1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
        self.bubble_sort_helper(nums, n-1)

    def bubble_sort(self, nums):
        n = len(nums)
        self.bubble_sort_helper(nums, n)
        return nums
        
```

```python
sol1 = Solution_R()
ar = [7,4,1,5,3,7]
sol1.bubble_sort(ar)
print(ar)
```
