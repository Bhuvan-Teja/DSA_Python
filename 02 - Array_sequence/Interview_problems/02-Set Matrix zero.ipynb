# Set Matrix Zero
## Problem
Problem Statement: Given a matrix if an element in the matrix is 0 then you will have to set its entire column and row to 0 and then return the matrix..

Examples\

Input: matrix=[[1,1,1],[1,0,1],[1,1,1]]\
Output: [[1,0,1],[0,0,0],[1,0,1]]\
Explanation: Since matrix[2][2]=0.\
Therfore the 2nd column and 2nd row wil be set to 0


Input: matrix=[[0,1,2,0],[3,4,5,2],[1,3,1,5]]\
Output:[[0,0,0,0],[0,4,5,0],[0,3,1,0]]\
Explanation:Since matrix[0][0]=0 and matrix[0][3]=0.\
Therefore 1st row, 1st column and 4th column will be set to 0

## Solution 


### Brute Force 

- Traverse the entire matrix.
- If an element is zero:
    - Mark all elements in its row (except already zero) as -1.
    - Mark all elements in its column (except already zero) as -1.
- Once the full traversal is complete, replace all -1 with 0.

```python
class Solution():
    def setzero(self, matrix):
        m = len(matrix)
        n = len(matrix[0])
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    print("zero position is :", i, j)
                    for col in range(n):
                        if matrix[i][col] != 0:
                            matrix[i][col] = -1

                    for row in range(m):
                        if matrix[row][j] != 0:
                            matrix[row][j] = -1

                    
        
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == -1:
                    matrix[i][j] = 0
                print(matrix[i][j])
```

```python
test = Solution()
matrix = [[1,1,1],[1,0,1],[1,1,1]]
test.setzero(matrix)
```

The complexity of the brute force method is O(M * N * (M+N)) = O(M*N)
Which is not a good solution as the input scales
