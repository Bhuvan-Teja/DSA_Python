#### Largest Element
- Given an array of integers nums, return the value of the largest element in the array

- Example 1\
    Input: nums = [3, 3, 6, 1]\
    Output: 6\
    Explanation: The largest element in array is 6

- Example 2\
    Input: nums = [3, 3, 0, 99, -40]\
    Output: 99\
    Explanation: The largest element in array is 99

```python
class Solution:
    def largestElement(self, nums):
        large = nums[0]
        for i in range(1, len(nums)):
            if nums[i] > large:
                large = nums[i]
        return large
```

```python
sol = Solution()
ar = [3,3,6,1]
print(sol.largestElement(ar))
```

#### Second Largest Element
- Given an array of integers nums, return the second-largest element in the array. If the second-largest element does not exist, return -1.
- Example 1\
    Input: nums = [8, 8, 7, 6, 5]\
    Output: 7
    - Explanation:\
    The largest value in nums is 8, the second largest is 7

- Example 2\
    Input: nums = [10, 10, 10, 10, 10]\
    Output: -1
    - Explanation:\
    The only value in nums is 10, so there is no second largest value, thus -1 is returned

- Constrains\
    1 <= nums.length <= 105\
    $-10^4$ <= nums[i] <= $10^4$\
    nums may contain duplicate elements.

```python
class Solution:
    def secondLargestElement(self, nums):
        largest = nums[0]
        secondLargest = -10**5 # we need to max -ve number 
        for i in range(len(nums)):
            if nums[i]>largest:
                secondLargest = largest
                largest = nums[i]
            elif nums[i]<largest and nums[i]>secondLargest:
                secondLargest = nums[i]
        return -1 if secondLargest == -10**5 else secondLargest
```

```python
sol = Solution()
ar = [10, 10, 10, 10, 10]
print(sol.secondLargestElement(ar))
```

#### Check if the Array is Sorted II [Leetcode 1752]
- Given an array nums of n integers, return true if the array nums is sorted in non-decreasing order or else false.
- Example 1\
    Input : nums = [1, 2, 3, 4, 5]\
    Output : true\
    Explanation : For all i (1 <= i <= 4) it holds nums[i] <= nums[i+1], hence it is sorted and we return true.

- Example 2\
    Input : nums = [1, 2, 1, 4, 5]\
    Output : false\
    Explanation : For i == 2 it does not hold nums[i] <= nums[i+1], hence it is not sorted and we return false.

- Constraints\
    1 <= n <= 100\
    1 <= nums[i] <= 100

```python
class Solution: # time complexity = O(n)
    def isSorted(self, nums):
        for i in range(1, len(nums)):
            if nums[i]<nums[i-1]:
                return False
        return True
```

```python
sol = Solution()
ar = [1,2,1,4,5]
print(sol.isSorted(ar))
```

#### Remove duplicates from sorted array [Leetcode 26]
- Given an integer array nums sorted in non-decreasing order, remove all duplicates in-place so that each unique element appears only once.\
    Return the number of unique elements in the array.\
    If the number of unique elements be k, then,\
    Change the array nums such that the first k elements of nums contain the unique values in the order that they were present originally.\
    The remaining elements, as well as the size of the array does not matter in terms of correctness.\
    The driver code will assess correctness by printing and checking only the first k elements of the modified array.\
    An array sorted in non-decreasing order is an array where every element to the right of an element is either equal to or greater in value than that element.

- Example 1\
    Input: nums = [0, 0, 3, 3, 5, 6]\
    Output: 4\
    Explanation:\
    Resulting array = [0, 3, 5, 6, _, _]\
    There are 4 distinct elements in nums and the elements marked as _ can have any value.

- Example 2\
    Input: nums = [-2, 2, 4, 4, 4, 4, 5, 5]\
    Output: 4\
    Explanation:\
    Resulting array = [-2, 2, 4, 5, _, _, _, _]\
    There are 4 distinct elements in nums and the elements marked as _ can have any value.

- Constraints\
    1 <= nums.length <= $10^5$\
    $-10^4$ <= nums[i] <= $10^4$\
    nums is sorted in non-decreasing order.

##### Brute Force Method!
- This method uses the set dataset to find the unique elements.

```python
class Solution:
    # Remember that python set will use unordered set, it does not store the order 
    # To use ordered set you can use the 3rd party library like ordered_Set
    # from ordered_set import OrderedSet
    def removeDuplicates(self, nums: list[int]) -> int:
        temp_set = set()
        for i in range(len(nums)):  
            temp_set.add(nums[i])
        j = 0
        for i in temp_set:
            nums[j] = i
            j+=1
        return j
```

```python
sol = Solution()
ar = [-2, 2, 4, 4, 4, 4, 5, 5]
index = sol.removeDuplicates(ar)
print(ar[:index])
print(index)
```

##### Optimal solution
- For this you will be using the 2 pointer approch
- where you place the 2 points in at 0th index
- check if i and j are pointing to the same value
- if yes move the j pointer to the right untill you find different value
- once j is at different value, assign the j pointed value to the i+1 position
- may be increment the i and assign the j to i
- repeat the steps  

```python
class Solution:
    def removeDuplicates_optimal(self, nums: list[int]) -> int:
        i = 0; j = 1
        while j<len(nums):
            if nums[i] == nums[j]:
                j+=1
            else:
                i+=1
                nums[i] = nums[j]
                j+=1
        return i+1
    
```

```python
opsol = Solution()
ar = [0,0,3,3,5,6]
index = opsol.removeDuplicates_optimal(ar)
print(index)
print(ar[:index])
```

#### Left Rotate Array by One
- Given an integer array nums, rotate the array to the left by one.\
    Note: There is no need to return anything, just modify the given array.

- Example 1\
    Input: nums = [1, 2, 3, 4, 5]\
    Output: [2, 3, 4, 5, 1]\
    Explanation:\
    Initially, nums = [1, 2, 3, 4, 5]\
    Rotating once to left -> nums = [2, 3, 4, 5, 1]

- Example 2\
    Input: nums = [-1, 0, 3, 6]\
    Output: [0, 3, 6, -1]\
    Explanation:\
    Initially, nums = [-1, 0, 3, 6]\
    Rotating once to left -> nums = [0, 3, 6, -1]

- Constraints\
    1 <= nums.length <= $10^5$\
    $-10^4$ <= nums[i] <= $10^4$

```python
class Solution:
    def rotateArrayByOne(self, nums):
        n = len(nums)
        temp = nums[0]
        for i in range(1, n):
            nums[i-1] = nums[i]
        nums[n-1] = temp
        return nums
```

```python
sol = Solution()
ar = [-1,0,3,6]
print(sol.rotateArrayByOne(ar))
```

#### Left Rotate Array by K Places
- Given an integer array nums and a non-negative integer k, rotate the array to the left by k steps.

- Example 1\
    Input: nums = [1, 2, 3, 4, 5, 6], k = 2\
    Output: nums = [3, 4, 5, 6, 1, 2]\
    Explanation:\
    rotate 1 step to the left: [2, 3, 4, 5, 6, 1]\
    rotate 2 steps to the left: [3, 4, 5, 6, 1, 2]

- Constraints\
    1 <= nums.length <= $10^5$\
    $-10^4$ <= nums[i] <= $10^4$\
    0 <= k <= $10^5$

##### Brute Force
- In this approch we are moving the k elements to the tmp array
- next the are left shifting them to the 0th index from kth position
- next we are placing the tmp elements to the last 
- Time complexity is O(K + N-K + )

```python
class Solution:
    def rotateArray(self, nums, k: int) -> None:
        n = len(nums)
        tmp = [0] * k
        j = 0
        for i in range(k):
            tmp[i] = nums[i]
        for i in range(k,n):
            nums[i-k] = nums[i]
        for i in range(n-k, n):
            nums[i] = tmp[j]
            j+=1
        return nums
```

```python
sol = Solution()
ar = [1,2,3,4,5,6]
k = 4
sol.rotateArray(ar, k)
print(ar)
```

##### Optimal Solution 
- Here we use 3 calls to the helper method which will reverse the array from start to end inplace
- 1st call is from 0 to k-1
- 2nd call is from k to n
- 3rd call is from 0 to n

```python
class Solution:
    def rotate_helper(self, nums, start, end):
        while start<end:
            nums[start], nums[end] = nums[end], nums[start]
            start+=1
            end-=1

    def rotateArray_op(self, nums, k: int) -> None:
        n = len(nums)
        k = k % n
        self.rotate_helper(nums, 0, k-1)
        self.rotate_helper(nums, k, n-1)
        self.rotate_helper(nums, 0, n-1)
        return nums
```

```python
sol = Solution()
ar = [1,2,3,4,5,6,7]
k = 8
sol.rotateArray_op(ar, k)
print(ar)
```
