```python

def large_num(nums):
    large = -10**6
    for i in nums:
        if i>large:
            large = i
    return large
```

```python
print(large_num(nums = [3, 3, 6, 1]))
```

```python
def reverse_integer(n):
    
    sign = -1 if n < 0 else 1
    reversed_val = int(str(abs(n))[::-1])
    
    return sign * reversed_val
```

```python
def selection_sort(nums): 
# logic is to find minimum of the array and swap to the 0th index of the array
    for i in range(len(nums)-1): #i=0 to n-2
        mini = i
        for j in range(i, len(nums)): # j = i to n-1
            if nums[j]<nums[mini]:
                mini = j
        nums[i], nums[mini] = nums[mini], nums[i]


    return nums
```

```python
def bubble_sort(nums):
# move max of 2 element and swap to the right, which will end up moving to the last index
    for i in range(len(nums)-1, -1, -1):
        for j in range(i):
            if nums[j]>nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums
```

```python
def insertion_sort(nums):
    for i in range(len(nums)):
        j = i
        while(j>0 and nums[j-1]>nums[j]):
            nums[j-1], nums[j] = nums[j], nums[j-1]
            j-=1

    return nums
```

```python
def second_large_brute(ar): # time complexity is O(NLogN + N)
    new_ar = sorted(ar)
    max2 = -1
    max = new_ar[len(new_ar)-1]
    for i in range(len(new_ar)-2, -1, -1):
        if new_ar[i] < max:
            max2 = new_ar[i]
            break
    
    return max2
```

```python
def second_large_better(ar): # Time complexity is O(N+N) = O(2N)
    max = ar[0]
    max2 = -1
    for i in range(1, len(ar)):
        if ar[i] > max:
            max = ar[i]
    for j in range(len(ar)):
        if ar[j]>max2 and ar[j]<max:
            max2 = ar[j]
    return max2
```

```python
def second_large_optimal(ar):
    max = ar[0]
    max2 = -10**5
    for i in range(1, len(ar)):
        if ar[i]>max:
            max2 = max
            max = ar[i]
        elif ar[i]<max and ar[i]>max2:
            max2 = ar[i]

    if max2 == -10**5:
        return -1
    else:
        return max2
```

```python
a = [8, 8, 7, 6, 5]
print(second_large_better(a))
print(second_large_optimal([1000,1000]))
```
